```python
import matplotlib.pyplot as plt
import numpy as np
```

# Worldwide

```python
# Labels
labels = [
    'FRP-64d', 'FRP-128d', 'FRP-256d', 'FRP-RF1', 'FRP-RF2', 'FRP-Alt1', 'FRP-Alt2', 'FRP-RF3',
    'N2V-BFS', 'N2V-DFS', 'N2V-Bal', 'N2V-Walks', 'N2V-HiDim'
]

# Corresponding test scores
test_scores = [
    0.5666, 0.5639, 0.5639, 0.4872, 0.4872, 0.5639, 0.5639, 0.4872, # FastRP
    0.5747, 0.5747, 0.5747, 0.5747, 0.5747 # Node2vec
]

# Model groupings for color coding (0 = FastRP, 1 = Node2Vec)
model_group = [0]*8 + [1]*5
colors = ['#1f77b4' if m == 0 else '#ff7f0e' for m in model_group]  # Blue for FastRP, Orange for Node2Vec

# Plot
plt.figure(figsize=(14, 6))
bars = plt.bar(labels, test_scores, color=colors)

# Add value labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.005, f'{yval:.3f}', ha='center', va='bottom', fontsize=8)

# Styling
plt.title('Worldwide - Model Performance Comparison (Test Scores)', fontsize=14)
plt.ylabel('Test Score (Accuracy)')
plt.xticks(rotation=45, ha='right')
plt.ylim(0.45, 0.60)
plt.legend(handles=[
    plt.Rectangle((0,0),1,1,color='#1f77b4', label='FastRP'),
    plt.Rectangle((0,0),1,1,color='#ff7f0e', label='Node2Vec')
], loc='lower right')

plt.tight_layout()
plt.savefig('model_performance_comparison.png', dpi=300)
plt.show()
```
