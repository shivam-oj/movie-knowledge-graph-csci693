```python
import matplotlib.pyplot as plt

# Common settings
labels = ['Flop', 'Moderate', 'Hit']
colors = ['#ff9999', '#66b3ff', '#99ff99']
explode = (0.05, 0.05, 0.05)

# Market-wise data
worldwide = [519, 542, 1413]
international = [1232, 577, 665]
domestic = [1094, 716, 664]

# Create figure and subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Pie chart for Worldwide
axs[0].pie(worldwide, labels=labels, autopct='%1.1f%%', startangle=140,
           colors=colors, explode=explode, shadow=True)
axs[0].set_title('Worldwide ROI Class Distribution (N=2474)')

# Pie chart for International
axs[1].pie(international, labels=labels, autopct='%1.1f%%', startangle=140,
           colors=colors, explode=explode, shadow=True)
axs[1].set_title('International ROI Class Distribution (N=2474)')

# Pie chart for Domestic
axs[2].pie(domestic, labels=labels, autopct='%1.1f%%', startangle=140,
           colors=colors, explode=explode, shadow=True)
axs[2].set_title('Domestic ROI Class Distribution (N=2474)')

# Keep circles and tight layout
for ax in axs:
    ax.axis('equal')

plt.tight_layout()

# Save the figure
plt.savefig('roi_class_distribution.png', dpi=300, bbox_inches='tight')  # or .pdf, .svg

plt.show()
```
